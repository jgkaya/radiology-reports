```python
import os
import re
import glob
import pandas as pd
```

```python
folder_path = "C:/Users/sbk07/Desktop/MIMICVeriseti" # Klasorun path'i yazildi
pattern_path =folder_path + "/**/*.txt"  # kalıp deseni
```

```python
# RegEx ile pattern olusturuldu
pattern = r"EXAMINATION:\s*(\S.*)\s*INDICATION:\s*(\S.*)\s*TECHNIQUE:\s*(\S.*)\s*COMPARISON:\s*(\S.*)\s*FINDINGS:\s*(\S[\s\S]*?)\s*IMPRESSION:\s*(\S.*)"
```

```python
dfs = []
```

```python
for file_path in glob.glob(pattern_path, recursive=True):
    with open(file_path, "r") as f:
        content = f.read()
        match = re.search(pattern, content)
        if match:
            examination = match.group(1)
            indication = match.group(2)
            technique = match.group(3)
            comparison = match.group(4)
            findings = match.group(5)
            impression = match.group(6)
            data = {'Examination': [examination],
                    'Indication': [indication],
                    'Technique': [technique],
                    'Comparison': [comparison],
                    'Findings': [findings],
                    'Impression': [impression]}
            df = pd.DataFrame(data)
            dfs.append(df)
```

```python
df = pd.concat(dfs, ignore_index=True)
```
